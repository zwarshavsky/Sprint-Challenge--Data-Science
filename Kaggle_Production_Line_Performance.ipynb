```python
df_train_categorical = "./data/train_categorical.csv"
df_train_numeric = "./data/train_numeric.csv"
df_train_date = "./data/train_date.csv"

df_train_list = [df_train_categorical,df_train_numeric,df_train_date]
```

```python
df_train_list
```

```python
import csv
from random import random
import pandas
my_dfs = {}
for index,path in enumerate(df_train_list):
    with open(path, 'r') as fin:
        reader = csv.reader(fin)
        rows = [l for l in reader if random() >= 0.90]
        my_dfs[index] = pandas.DataFrame(rows)
```

```python
for i in range(len(my_dfs)):
    my_dfs[i].to_pickle(f"./{i}_train.pkl")
```
